```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
#exercicio1A

n =13 # numero de pontos

y = np.linspace(-2,2,n)
t = np.linspace(-2,2,n)
Y,T = np.meshgrid(y,t)
# Y,T posições
#dy,dt direção da seta
#plt.scatter(Y,T)

f = 2-Y 

angulo = np.arctan(f)
dy = np.cos(angulo)
dt = np.sin(angulo)
plt.quiver(T,Y,dy,dt)
```

```python
#Exercício 2

cs = [1, 15, 0.5, 7, 8]
x = sp.Symbol("x")
f = sp.Function("f")(x)
ed = sp.Eq(f.diff(),0.5*f-450)

sol = sp.dsolve(ed,f)
fd = sol.rhs

c1, _ = tuple(fd.free_symbols)

for i in cs:
  s = fd.subs(c1,i)
  sp.plot(s, show=True)
```
